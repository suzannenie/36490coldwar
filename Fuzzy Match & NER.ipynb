```python
import os
arr = os.listdir('data_txts')
print(arr)
text = []
for doc in arr:
    with open('data_txts/'+ doc) as f:
        text.append(f.readlines())
```

```python
nlp = en_core_web_sm.load()
```

```python
flat_text =  ' '.join([item for sublist in text for item in sublist])
flat_text
doc = nlp(flat_text)
#https://www.codespeedy.com/named-entity-recognition-using-spacy-in-python/
```

```python
# uncomment to see named entities and labels
print([(z.text, z.label_) for z in doc.ents])
displacy.render(doc,style="ent",jupyter=True)
```

```python
persons = [ent.text for ent in doc.ents if ent.label_ == 'PERSON']
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# https://stackoverflow.com/questions/35171710/how-to-group-words-whose-levenshtein-distance-is-more-than-80-percent-in-python

combined_list = persons
#print('input names:', combined_list)

grs = list() # groups of names with distance > 80
for name in combined_list:
    for g in grs:
        if all(fuzz.ratio(name, w) > 80 for w in g):
            g.append(name)
            break
    else:
        grs.append([name, ])

print('output groups:', grs)
#outlist = [el for g in grs for el in g]
#print('output list:', outlist)
```

```python
%pip install pandas
import pandas as pd
```

```python
import re
name_matches = grs
names = [re.sub(r'[^a-zA-Z ]', '', item[0].replace('\n', '')) for item in name_matches]
names
```

```python
same_names = []
for element in name_matches: 
    same_names.append(", ".join(element).replace('\n', ''))
same_names
```

```python
name_matches = pd.DataFrame(list(zip(names, same_names)),
              columns=['name', 'matches'])
pd.set_option('display.max_rows', None)
name_matches
```
